# VISUALIZATION USING DIFFERENT PLOTS

# SCATTER PLOT:

```python
# We will make a scatterplot of two of the four Iris features 


iris.plot(kind="scatter", x="SepalLengthCm", y= "SepalWidthCm")
```

# JOINT PLOT:

```python
## we can do it also with the seaborn library

## A seaborn jointplot shows bivariate scatterplots and univariate histograms in the 
# same figure

sns.jointplot( data=iris,x="SepalLengthCm", y="SepalWidthCm", size=5)
```

# DENSITY PLOT : KDE

```python
## we can do it also with the seaborn library

## A seaborn kdeplot shows density scatterplots and univariate histograms in the 
# same figure

sns.kdeplot(data= iris, x="SepalLengthCm", y="SepalWidthCm", size=5)
```

# FACET GRID:

```python
sns.FacetGrid(iris, hue="Species", palette="husl",size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

# BOXPLOT

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

# VIOLIN PLOT:

```python
sns.violinplot(data=iris, x="Species", y="PetalWidthCm",
                palette="rocket")
```

# PAIR PLOT:

```python
sns.pairplot(iris.drop("Id", axis=1), 
             hue ="Species",palette= "husl", size=3) 
 
```

```python
dataset.tail()
```

# HEATMAP

```python
plt.figure(figsize=(12,8)) 
sns.heatmap(dataset.corr(), annot=True, cmap='Dark2_r', linewidths = 2)
plt.show()
```

```python
plt.figure(figsize=(25,15))

plt.subplot(2,2,1)
sns.histplot(dataset['SepalLengthCm'], color = 'red', kde = True).set_title('SepalLengthCm Interval and Counts')

plt.subplot(2,2,2)
sns.histplot(dataset['SepalWidthCm'], color = 'green', kde = True).set_title('SepalWidthCm Interval and Counts')

plt.subplot(2,2,3)
sns.histplot(dataset['PetalLengthCm'], kde = True, color = 'blue').set_title('PetalLengthCm Interval and Counts')

plt.subplot(2,2,4)
sns.histplot(dataset['PetalWidthCm'], kde = True, color = 'black').set_title('PetalWidthCm Interval and Counts')
```

# DATA VISUALIZATION USING BAR

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.barplot(x = 'Species', y = 'SepalLengthCm', data = dataset, palette="cubehelix")
plt.subplot(2,2,2)
sns.barplot(x = 'Species', y = 'SepalWidthCm', data = dataset, palette="Oranges")
plt.subplot(2,2,3)
sns.barplot(x = 'Species', y = 'PetalLengthCm', data = dataset, palette="Oranges")
plt.subplot(2,2,4)
sns.barplot(x = 'Species', y = 'PetalWidthCm', data = dataset, palette="cubehelix")
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.distplot(dataset['SepalLengthCm'], color="red").set_title('SepalLength Interval')
plt.subplot(2,2,2)
sns.distplot(dataset['SepalWidthCm'], color="green").set_title('SepalWidth Interval')
plt.subplot(2,2,3)
sns.distplot(dataset['PetalLengthCm'], color="blue").set_title('PetalLength Interval')
plt.subplot(2,2,4)
sns.distplot(dataset['PetalWidthCm'], color="black").set_title('PetalWidth Interval')
```
